```python
#Dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
import pandas as pd 
charity_df = pd.read_csv("Resources/charity_data.csv")
charity_df.head()
```

```python
# To drop ID columns for , 'EIN' and 'NAME'.
charity_df = charity_df.drop(["EIN", "NAME"], axis = 1)
charity_df.head()
```

```python
#To check the number of unique values for each column
charity_df.nunique()
```

```python
# Application type value for binning
charity_df.APPLICATION_TYPE.value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = charity_df["APPLICATION_TYPE"].value_counts().loc[lambda x : x < 500].index


# Replace in dataframe
for app in application_types_to_replace:
    charity_df['APPLICATION_TYPE'] = charity_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
charity_df['APPLICATION_TYPE'].value_counts()
```

```python
#classification value counts for binning
charity_df["CLASSIFICATION"].value_counts()
```

```python
# Converting categorical data to numeric with `pd.get_dummies`
converted_df = pd.get_dummies(charity_df)
converted_df
```

```python
# Spliting the converted data 
y = converted_df["IS_SUCCESSFUL"].values
X = converted_df.drop(columns= "IS_SUCCESSFUL").values

# Split i into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```
